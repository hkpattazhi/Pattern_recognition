```python
covariance_matrix1, covariance_matrix2 = utils.generate_covariance_mat(a=3, b=4, c=6, alpha=.2, beta=0.1)
mean_vector1 = np.array([2,4,6])
mean_vector2 = np.array([2,-4,-6])
#generating 200 point for class1 and class2
data_gen = data_generation.data_generation()
random_gaussian_matrix1 = data_gen.gausian_random(200,3)
random_gaussian_matrix2 = data_gen.gausian_random(200,3)
matrix_train1= data_gen.generate_cov_data( random_gaussian_matrix1.T,covariance_matrix1, mean_vector1)
matrix_train2= data_gen.generate_cov_data( random_gaussian_matrix2.T,covariance_matrix2,mean_vector2)
# creating dataframe with 2 classes
df_train = create_df(matrix_train1, matrix_train2)
```

```python
#calculating the coefficients A,B,C for quadratic bayes optimal classifier
A, B, C = bayes_optimal.fit_model(matrix_train1, matrix_train2)
```

```python
def calc_prediction_score(x,y, A,B,C):
    return (A[0][0]*(x**2) +
           A[1][1]*(y**2) + 
           (A[0][1] +A[1][0])*x*y + 
           (2*A[0][2] + 2*A[2][0] + B[0])*x + 
           (2*A[1][2] + 2*A[2][1] +B[1])*y + 
           2*B[2] +A[2][2]*(4) +C)
                                                     
```
